```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/home/bethtian/fintech545/beth-fintech545/Project02/DailyPrices.csv")
df
```

```python
required_stocks = ['SPY', 'AAPL', 'EQIX']
stock_prices = df[required_stocks]  
stock_prices
```

# A.

```python
arithmetic_returns = stock_prices.pct_change()

arithmetic_returns_zero_mean = arithmetic_returns - arithmetic_returns.mean()

arithmetic_std = arithmetic_returns_zero_mean.std()

print("\nLast 5 rows of Arithmetic Returns (Zero Mean):")
print(arithmetic_returns_zero_mean.tail(5))
print("\nStandard Deviation of Arithmetic Returns (Zero Mean):")
print(arithmetic_std)
```

# B.

```python
log_returns = np.log(stock_prices / stock_prices.shift(1))

log_returns_zero_mean = log_returns - log_returns.mean()

log_std = log_returns_zero_mean.std()

print("\nLast 5 rows of Log Returns (Zero Mean):")
print(log_returns_zero_mean.tail(5))

print("\nStandard Deviation of Log Returns (Zero Mean):")
print(log_std)
```
